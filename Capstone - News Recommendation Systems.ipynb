````python
def decontracted(phrase):
    phrase = re.sub(r"won\'t", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase
````

````python
import re
````

````python
def news_recommendations(news_title,number_of_hits=10):
    # get the news row
    news = news_df.loc[news_df.title==news_title] 
    
    #split the keywords
    keywords = news['text_description_processed'].iloc[0].split() 

    #set the list of words to query_doc
    query_doc = keywords 
    
    # get a bag of words from the query_doc
    query_doc_bow = dictionary.doc2bow(query_doc) 

    #convert the regular bag of words model to a tf-idf model
    query_doc_tfidf = tfidf[query_doc_bow] 

    # get similarity values between input movie and all other movies
    similarity_array = sims[query_doc_tfidf] 

    #Convert to a Series
    similarity_series = pd.Series(similarity_array.tolist(), index=news_df.title.values) 

    #get the most similar movies 
    similarity_output = similarity_series.sort_values(ascending=False)
    
    recommended_news = similarity_series.sort_values(ascending=False)[1:number_of_hits+1] 
    #get the top matching results, i.e. most similar movies; start from index 1 because every movie is most similar to itself
    print("News Title : " + news_title)
    print()
    print("                      Top Recommended News article for along with Similarity Score :      ")
    print()
    print(recommended_news)
    
    
    recommended_news_df = pd.DataFrame(similarity_output)
    recommended_news_df = recommended_news_df.drop_duplicates() 
    recommended_news_df.reset_index(inplace=True)
    recommended_news_df.columns = ['title', 'score']
    #print()
    #print()
    #print(recommended_news_df.head())
    #print()
    #print()
    recommended_news_df['score_normalized'] = (recommended_news_df['score']-min(recommended_news_df['score'])) / (max(recommended_news_df['score']) - min(recommended_news_df['score']))
    #print()
    #print()
    #print(recommended_news_df[['title','score']])
    
    #return  recommended_news_df
    return similarity_output
````

#### --MAE <a class="anchor" id="mae"></a>

````python
def mae(prediction, ground_truth):

    prediction = prediction[ground_truth.nonzero()]
    
    ground_truth = ground_truth[ground_truth.nonzero()]
    
    return mean_absolute_error(prediction, ground_truth)
````

#### --RMSE <a class="anchor" id="rmse"></a>

````python
def rmse(prediction, ground_truth):
    prediction = prediction[ground_truth.nonzero()]
    ground_truth = ground_truth[ground_truth.nonzero()]
    return sqrt(mean_squared_error(prediction, ground_truth))
````
